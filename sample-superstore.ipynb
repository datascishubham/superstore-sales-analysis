```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
df= pd.read_csv('Superstore.csv', encoding_errors='replace')
```

```python
df.head()
```

```python
#understand the data
df.describe()
```

```python
df.info()
```

```python
df.columns
```

```python
# check the null 
df.isnull().sum()
```

```python
#check the duplicated values
df.duplicated().sum()
```

```python
# remove the duplicate value
df.drop_duplicates(inplace=True)
```

```python
# add new column
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce') 

df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Profit Margin'] = df['Profit'] / df['Sales']
```

```python
df.head()
```

```python
# find the top 10 product by sales
top_product= df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
top_product
```

```python
top_product.plot(kind='bar' ,figsize=(12,6))
plt.title("top 10 product by sales")
plt.xlabel("Product Name")
plt.ylabel("Sales")

plt.show()
```

```python
# create a barplot category by profit
sns.barplot(x="Category",y="Profit",data=df, estimator=sum)
plt.title("Profit by category")
```

```python
# find the monthly wise sale with dates
Monthliy_sales= df.groupby(pd.Grouper(key="Order Date", freq='ME'))['Sales'].sum()
```

```python
Monthliy_sales
```

```python
Monthliy_sales.plot(figsize=(12,6))
plt.title("Monthly sales trend")
```

```python
# find city wise sales
state_sales = df.groupby('State')['Sales'].sum().sort_values()
state_sales.plot(kind='barh', figsize=(12,8))
plt.title("State-wise Sales")
```

```python
corr = df[['Sales','Profit' ,'Discount','Quantity']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
segment_profit= df.groupby(['Segment'])['Profit'].sum()
```
